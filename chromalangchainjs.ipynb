```python
from langchain.vectorstores import Chroma
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import TextLoader
```

```python
# Load and process the text
loader = TextLoader('/content/appliedaiqa.txt')
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
texts = text_splitter.split_documents(documents)
```

```python
# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
embedding_function
```

```python
embedding_function.embed_documents(['this is an example embedding','secone'])
```

```python
len(texts)
```

```python
import chromadb
chroma_client = chromadb.Client()
```

```python

# load it into Chroma and look at embeddings
db = Chroma.from_documents(texts,embedding=embedding_function)
print(db._collection.get(include=['embeddings']))

```

```python
print(db._collection.get(ids=['59229a6c-7f3b-11ee-a585-0242ac1c000c'],include=['embeddings']))
```

```python
# query it
query = "what do you believe in"
docs = db.similarity_search(query)

# print results
print(docs[0].page_content)
```
